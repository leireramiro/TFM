## Load results

```python
# for result visualization
import numpy as np
import matplotlib.pyplot as plt

# for loading data
import pickle 
```

```python
# Loading MSE of all models

# Energy

# without data augmentatition
f_mse_energy_samples = open('mse_energy_samples.pickle', 'rb')
(sgdr_scores1_e,dt_scores1_e,svr_scores1_e) = pickle.load(f_mse_energy_samples)
f_mse_energy_samples.close()

# Mixup data augmentatition
f_mse_energy_samples_MUaug = open('mse_energy_samples_MUaug.pickle', 'rb')
(sgdr_scores2_MU_e,dt_scores2_MU_e,svr_scores2_MU_e) = pickle.load(f_mse_energy_samples_MUaug)
f_mse_energy_samples_MUaug.close()

f_mse_energy_samples_MUaug2 = open('mse_energy_samples_MUaug2.pickle', 'rb')
(sgdr_scores2_MU_e2,dt_scores2_MU_e2,svr_scores2_MU_e2) = pickle.load(f_mse_energy_samples_MUaug2)
f_mse_energy_samples_MUaug2.close()

f_mse_energy_samples_MUaug_l = open('mse_energy_samples_MUaug_l.pickle', 'rb')
(sgdr_scores_20_e,sgdr_scores_50_e,sgdr_scores_80_e,dt_scores_20_e,dt_scores_50_e,dt_scores_80_e,svr_scores_20_e,svr_scores_50_e,svr_scores_80_e) = pickle.load(f_mse_energy_samples_MUaug_l)
f_mse_energy_samples_MUaug_l.close()

# Noise injection data augmentatition
f_mse_energy_samples_Naug = open('mse_energy_samples_Naug.pickle', 'rb')
(sgdr_scores2_N_e,dt_scores2_N_e,svr_scores2_N_e) = pickle.load(f_mse_energy_samples_Naug)
f_mse_energy_samples_Naug.close()

f_mse_energy_samples_Naug2 = open('mse_energy_samples_Naug2.pickle', 'rb')
(sgdr_scores2_N_e2,dt_scores2_N_e2,svr_scores2_N_e2) = pickle.load(f_mse_energy_samples_Naug2)
f_mse_energy_samples_Naug2.close()

f_mse_energy_samples_NIaug_n = open('mse_energy_samples_NIaug_n.pickle', 'rb')
(sgdr_scores_20_e2,sgdr_scores_50_e2,sgdr_scores_80_e2,dt_scores_20_e2,dt_scores_50_e2,dt_scores_80_e2,svr_scores_20_e2,svr_scores_50_e2,svr_scores_80_e2) = pickle.load(f_mse_energy_samples_NIaug_n)
f_mse_energy_samples_NIaug_n.close()

# Valance

# without data augmentatition
f_mse_valence_samples = open('mse_valence_samples.pickle', 'rb')
(sgdr_scores1_v,dt_scores1_v,svr_scores1_v) = pickle.load(f_mse_valence_samples)
f_mse_valence_samples.close()

# Mixup data augmentatition
f_mse_valence_samples_MUaug = open('mse_valence_samples_MUaug.pickle', 'rb')
(sgdr_scores2_MU_v,dt_scores2_MU_v,svr_scores2_MU_v) = pickle.load(f_mse_valence_samples_MUaug)
f_mse_valence_samples_MUaug.close()

f_mse_valence_samples_MUaug2 = open('mse_valence_samples_MUaug2.pickle', 'rb')
(sgdr_scores2_MU_v2,dt_scores2_MU_v2,svr_scores2_MU_v2) = pickle.load(f_mse_valence_samples_MUaug2)
f_mse_valence_samples_MUaug2.close()

f_mse_valence_samples_MUaug_l = open('mse_valence_samples_MUaug_l.pickle', 'rb')
(sgdr_scores_20_v,sgdr_scores_50_v,sgdr_scores_80_v,dt_scores_20_v,dt_scores_50_v,dt_scores_80_v,svr_scores_20_v,svr_scores_50_v,svr_scores_80_v) = pickle.load(f_mse_valence_samples_MUaug_l)
f_mse_valence_samples_MUaug_l.close()

# Noise injection data augmentatition
f_mse_valence_samples_Naug = open('mse_valence_samples_Naug.pickle', 'rb')
(sgdr_scores2_N_v,dt_scores2_N_v,svr_scores2_N_v) = pickle.load(f_mse_valence_samples_Naug)
f_mse_valence_samples_Naug.close()

f_mse_valence_samples_Naug2 = open('mse_valence_samples_Naug2.pickle', 'rb')
(sgdr_scores2_N_v2,dt_scores2_N_v2,svr_scores2_N_v2) = pickle.load(f_mse_valence_samples_Naug2)
f_mse_valence_samples_Naug2.close()

f_mse_valence_samples_NIaug_n = open('mse_valence_samples_NIaug_n.pickle', 'rb')
(sgdr_scores_20_v2,sgdr_scores_50_v2,sgdr_scores_80_v2,dt_scores_20_v2,dt_scores_50_v2,dt_scores_80_v2,svr_scores_20_v2,svr_scores_50_v2,svr_scores_80_v2) = pickle.load(f_mse_valence_samples_NIaug_n)
f_mse_valence_samples_NIaug_n.close()
```

## Predicting energy

### Default data augmentation hyperparameters

```python
# range of the train set 
x_range = [item for item in list(range(10,110,10))]

f, axes = plt.subplots(nrows=1, ncols=3,sharey=True, figsize=(10, 5))

# creating the subplots
axes[0].plot(x_range, sgdr_scores1_e, '-o', label='without data aug.')
axes[0].plot(x_range, sgdr_scores2_MU_e, '-o', label='with MIXUP (=0.5)')
#axes[0].plot(x_range, sgdr_scores2_MU_e2, '-o', label='with MIXUP (=0.25)')
axes[0].plot(x_range, sgdr_scores2_N_e, '-o', label='with Noise injection (n=5)')
axes[0].set_xlabel('% train data', fontsize = 15)
axes[0].set_ylabel('MSE', fontsize = 15)

axes[1].plot(x_range, dt_scores1_e, '-o', label='without data aug.')
axes[1].plot(x_range, dt_scores2_MU_e, '-o', label='with MIXUP (=0.5)')
#axes[1].plot(x_range, dt_scores2_MU_e2, '-o', label='with MIXUP (=0.25)')
axes[1].plot(x_range, dt_scores2_N_e, '-o', label='with Noise injection (n=5)')
axes[1].set_xlabel('% train data', fontsize = 15)

axes[2].plot(x_range, svr_scores1_e, '-o', label='without data aug.')
axes[2].plot(x_range, svr_scores2_MU_e, '-o', label='with MIXUP (=0.5)')
#axes[2].plot(x_range, svr_scores2_MU_e2, '-o', label='with MIXUP (=0.25)')
axes[2].plot(x_range, svr_scores2_N_e, '-o', label='with Noise injection (n=5)')
axes[2].set_xlabel('% train data', fontsize = 15)

# set the general title
f.suptitle('MSE results for energy score', fontsize=20) 

# set the legend
labels = ["without data aug.", "with MIXUP (=0.5)", "with Noise injection (n=5)"]
f.legend(labels=labels,loc="upper left",bbox_to_anchor=(1, 1), fontsize = 15)

# set the title to subplots
axes[0].set_title("SGD Regression", size = 15)
axes[1].set_title("DT Regression", size = 15)
axes[2].set_title("SV Regression", size = 15)

# set spacing
f.tight_layout()

plt.show()
```
